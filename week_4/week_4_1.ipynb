# <span style="color:#003366;">بررسی IPv4، IPv6، مفهوم NAT و وضعیت پیاده‌سازی IPv6 در ایران</span>

<div dir="rtl" align="right">  
پروتکل اینترنت (Internet Protocol) یا به اختصار IP یکی از مهم‌ترین اجزای شبکه‌های مخابراتی و اینترنت است که وظیفه‌ی آدرس‌دهی و مسیریابی داده‌ها در شبکه را برعهده دارد.
</div>

<div dir="rtl" align="right">  
هر دستگاهی که به شبکه متصل می‌شود (مثل گوشی، مودم، سرور و...) برای شناسایی باید یک آدرس IP داشته باشد.  
در طول زمان دو نسخه‌ی اصلی از این پروتکل به‌کار رفته‌اند: IPv4 و IPv6.  
</div>

<div dir="rtl" align="right"> 
با افزایش سریع کاربران اینترنت و دستگاه‌های متصل، محدودیت‌های IPv4 باعث شد نسخه‌ی جدیدتر یعنی IPv6 طراحی و معرفی شود.
</div>


## <span style="color:#003366;">IPv4</span>
<div dir="rtl" align="right"> 
IPv4 (Internet Protocol version 4) قدیمی‌ترین و پرکاربردترین نسخه‌ی پروتکل اینترنت است که از سال 1981 معرفی و تا امروز در بیشتر شبکه‌ها استفاده می‌شود.

### <span style="color:#0099FF;">ساختار IPv4 :</span>
 
- طول آدرس ← 32 بیت 
- به‌صورت چهار عدد جداشده با نقطه (.) نوشته می‌شود، مانند:  
  `192.168.1.10`
- مجموعاً حدود 4.3 میلیارد آدرس یکتا (2³²) را فراهم می‌کند.


### <span style="color:#0099FF;">ویژگی‌ها:</span>
1. پشتیبانی گسترده توسط همه‌ی تجهیزات و نرم‌افزارها
2. ساختار ساده و قابل درک برای انسان 
3. استفاده از روش‌های سنتی مسیریابی (Routing)


### <span style="color:green;">مزایای IPv4 :</span>

1. سازگاری بالا با اغلب تجهیزات شبکه و نرم‌افزارها  
2. پایداری و تجربه‌ی عملی طولانی‌مدت (بیش از ۴ دهه استفاده)  
3. پشتیبانی قوی از NAT برای صرفه‌جویی در آدرس‌ها  
4. ساختار ساده‌تر برای پیاده‌سازی  

### <span style="color:red;">معایب IPv4 :</span>

1. محدود بودن تعداد آدرس‌ها (فقط 4.3 میلیارد)  
2. نیاز به NAT برای اشتراک‌گذاری آدرس‌ها  
3. نبود امنیت ذاتی (نیاز به IPsec به‌صورت افزونه)  
4. پشتیبانی ضعیف‌تر از QoS (کیفیت سرویس)  
5. افزایش پیچیدگی در مدیریت آدرس‌ها به دلیل کمبود آدرس عمومی  

---

## <span style="color:#003366;">IPv6</span>

<div dir="rtl" align="right"> 
IPv6 (Internet Protocol version 6) جدیدترین نسخه‌ی IP است که توسط IETF برای حل محدودیت‌های IPv4 طراحی شد.
</div>

### <span style="color:#0099FF;">ساختار IPv6 :</span>

- طول آدرس ← 128 بیت  
- به‌صورت هشت بخش هگزادسیمال جداشده با «:» نمایش داده می‌شود، مانند:  
  `2001:0db8:85a3:0000:0000:8a2e:0370:7334`  
- تعداد آدرس‌ها ← حدود 3.4 × 10³⁸ (تقریباً بی‌نهایت برای کاربردهای کنونی)

### <span style="color:#0099FF;">ویژگی‌ها:</span>

1. دارای ساختار سلسله‌مراتبی و خودکار در آدرس‌دهی  
2. پشتیبانی بومی از IPsec (امنیت در سطح پروتکل)  
3. قابلیت پیکربندی خودکار (Auto-Configuration)  
4. ساده‌تر شدن سربار پردازش در روترها  

### <span style="color:green;">مزایای IPv6 :</span>

1. تعداد نامحدود آدرس‌ها برای تمام دستگاه‌های دنیا  
2. امنیت درونی بالا به‌دلیل وجود IPsec  
3. پشتیبانی از پیکربندی خودکار (Stateless Address Auto Configuration)  
4. بهبود در عملکرد QoS برای سرویس‌هایی مانند ویدیو و VoIP  
5. عدم نیاز به NAT (هر دستگاه می‌تواند آدرس عمومی یکتا داشته باشد)  
6. بهینه‌سازی در ساختار Header و کاهش بار پردازشی مسیریاب‌ها  

### <span style="color:red;">معایب IPv6 :</span>

1. عدم سازگاری مستقیم با IPv4 (نیاز به Gateway یا Dual Stack)  
2. هزینه بالای ارتقاء تجهیزات و نرم‌افزارها  
3. کمبود نیروی متخصص و تجربه عملی در برخی کشورها  
4. پشتیبانی ناقص در برخی شبکه‌های قدیمی یا ISPهای کوچک  


### <span style="color:#0099FF;">کاربردهای واقعی IPv6 :</span>

- اینترنت اشیاء (IoT): میلیاردها حسگر و دستگاه به آدرس‌های یکتا نیاز دارند.  
- شبکه‌های 5G و 6G: طراحی این شبکه‌ها بر پایه IPv6 انجام شده است.  
- امنیت سایبری: IPsec ذاتی در IPv6 امنیت ارتباطات را افزایش می‌دهد.  
- مراکز داده و Cloud: استفاده از IPv6 باعث بهبود مقیاس‌پذیری و تخصیص پویا در ماشین‌های مجازی می‌شود.  



| ویژگی | IPv4 | IPv6 |
|:--:|:--:|:--:|
| طول آدرس | 32 بیت | 128 بیت |
| نمایش آدرس | 192.168.1.1 | 2001:db8::1 |
| تعداد آدرس‌ها | 4.3 میلیارد | تقریباً بی‌نهایت |
| امنیت (IPsec) | افزونه اختیاری | اجباری و داخلی |
| پیکربندی خودکار | ندارد | دارد |
| استفاده از NAT | بله (ضروری) | خیر (غیرضروری) |
| کارایی در QoS | محدود | بهبود یافته |
| سازگاری | بسیار بالا | در حال رشد |

## <span style="color:#003366;">NAT</span>

<div dir="rtl" align="right"> 
NAT (Network Address Translation) فناوری‌ای است که برای رفع مشکل کمبود آدرس در IPv4 استفاده می‌شود.  
NAT آدرس‌های خصوصی (Private IP) در شبکه داخلی را به یک یا چند آدرس عمومی (Public IP) ترجمه می‌کند تا دستگاه‌های داخلی بتوانند به اینترنت دسترسی پیدا کنند.  
</div> 

<div dir="rtl" align="right"> 
مثلاً در یک شبکه خانگی:  
هر دستگاه (موبایل، لپ‌تاپ، تلویزیون) دارای آدرس داخلی مانند 192.168.1.x است.  
مودم با استفاده از NAT همه‌ی درخواست‌ها را با یک آدرس عمومی مثلاً 85.185.100.12 به اینترنت می‌فرستد.
</div>

### <span style="color:green;">مزایا:</span>

1. صرفه‌جویی در مصرف آدرس‌های IPv4  
2. افزایش امنیت شبکه داخلی (آدرس‌ها پنهان می‌شوند)  
3. امکان اتصال چند دستگاه به یک آدرس عمومی  

### <span style="color:red;">معایب:</span>

1. افزایش تأخیر (Latency) در ارتباطات به دلیل ترجمه‌ی آدرس‌ها  
2. مشکل در برقراری برخی ارتباطات نظیر VoIP، VPN یا P2P  
3. کاهش شفافیت end-to-end (برخلاف اصل طراحی اینترنت)  

<div dir="rtl" align="right"> 
در IPv6 به دلیل فراوانی آدرس‌ها دیگر نیازی به NAT نیست، و ارتباط end-to-end واقعی برقرار می‌شود.
</div>

## <span style="color:#003366;">وضعیت استفاده از IPv6 در ایران</span>

<div dir="rtl" align="right"> 
با وجود معرفی IPv6 از سال 1999، پیاده‌سازی کامل آن در ایران هنوز محدود است.  
برخی شرکت‌های بزرگ و دانشگاه‌ها مثل زیرساخت، دانشگاه تهران، و برخی ISPها IPv6 را آزمایش یا به‌صورت محدود فعال کرده‌اند.  
اما در سطح کاربران خانگی هنوز بیشتر شبکه‌ها روی IPv4 و NAT کار می‌کنند.  
</div>

<div dir="rtl" align="right"> 
طبق آمار جهانی Google IPv6 Stats، ضریب نفوذ IPv6 در ایران کمتر از 5٪ است (تا سال 2025).
</div>
    
### <span style="color:#0099FF;">دلایل کندی مهاجرت:</span>

1. هزینه بالای ارتقاء تجهیزات شبکه و مودم‌ها  
2. نبود آموزش و نیروی متخصص کافی در حوزه IPv6  
3. پشتیبانی ناقص در شبکه‌های قدیمی و زیرساخت‌های مخابراتی  
4. اتکا به NAT و IPv4 خصوصی به‌عنوان راه‌حل موقت  

### <span style="color:#0099FF;">چشم‌انداز آینده:</span>
<div dir="rtl" align="right"> 
با گسترش IoT و شبکه‌های 5G و 6G، نیاز به IPv6 اجتناب‌ناپذیر است.  
وزارت ارتباطات و شرکت زیرساخت پروژه‌هایی برای توسعه تدریجی IPv6 در شبکه ملی اطلاعات آغاز کرده‌اند.  
پیش‌بینی می‌شود در ۵ تا ۱۰ سال آینده استفاده از IPv6 در ایران رشد قابل توجهی داشته باشد.  
</div>
