# بررسی تورم قیمت اینترنت و خدمات مخابراتی در مقایسه با تورم عمومی در ده سال اخیر ایران

## مقدمه
<pre style="white-space: pre-wrap; font-family:-apple-system, BlinkMacSystemFont,'Segoe UI',Roboto,'Noto Sans','Helvetica Neue',Arial;">
تحولات بخش فناوری اطلاعات و ارتباطات (ICT) نقش مهمی در رشد اقتصادی کشورها دارد. در ایران طی دهه اخیر مصرف دیتای کاربران رشد چشمگیری داشته و اینترنت به یک خدمت ضروری تبدیل شده است. از سوی دیگر، اقتصاد کشور با تورم بالا، کاهش ارزش پول ملی و افزایش هزینه‌های ارزی مواجه بوده که هم بر هزینه‌های کاربران و هم بر هزینه‌های عملیاتی اپراتورها اثر گذاشته است.

با وجود آنکه برای بسیاری از کالاها و خدمات، شاخص‌های رسمی قیمت به صورت سالانه منتشر می‌شود، اما در حوزهٔ اینترنت و مخابرات، شاخص مستقلی وجود ندارد. به همین دلیل روند قیمتی اینترنت باید از گزارش‌های افزایش تعرفه‌ها و آمارهای رسمی تورم عمومی استنتاج شود.
</pre>

## وضعیت تورم عمومی ایران در ده سال اخیر
<pre style="white-space: pre-wrap; font-family:-apple-system, BlinkMacSystemFont,'Segoe UI',Roboto,'Noto Sans','Helvetica Neue',Arial;">
بر اساس گزارش‌های رسمی بانک مرکزی و مرکز آمار، نرخ تورم ایران از سال‌های میانی دهه ۱۳۹۰ به بعد روندی صعودی و پرنوسان داشته و در چند سال اخیر رکوردهای بالایی ثبت کرده است. در بازهٔ حدوداً ۱۳۹۳ تا ۱۴۰۳:
تورم سالانه در برخی سال‌ها کمتر از ۱۰٪ بوده (۱۳۹۵–۱۳۹۶)
اما از ۱۳۹۷ به بعد تورم سالانه به محدوده ۳۰٪ تا ۵۰٪ رسیده
مجموع افزایش سطح عمومی قیمت‌ها طی این دوره چند صد درصد بوده است
به عبارت دیگر، در ده سال اخیر قیمت اغلب کالاها و خدمات نسبت به ابتدای دوره چندین برابر شده‌اند. 
</pre>

## روند قیمت اینترنت و خدمات مخابراتی
<pre style="white-space: pre-wrap; font-family:-apple-system, BlinkMacSystemFont,'Segoe UI',Roboto,'Noto Sans','Helvetica Neue',Arial;">
۱) ویژگی کلیدی: افزایش پله‌ای

برخلاف کالاهایی مثل خوراک یا مسکن که «ماهانه» تغییر قیمت می‌دهند، اینترنت معمولاً در ایران به شکل زیر افزایش قیمت داشته است:
چند سال بدون تغییر نگه‌داشته می‌شود سپس در یک مقطع مشخص یک «پرش قیمتی» دارد
این پرش‌ها معمولاً به دلیل سیاست‌گذاری رگولاتوری، فشار هزینه‌های اپراتورها یا تغییر تعرفه‌ها اتفاق می‌افتد.

۲) نمونه‌های مهم افزایش قیمت در ده سال اخیر
</pre>

## بر اساس گزارش‌های رسمی و اطلاعیه‌های اپراتورها: 
<pre style="white-space: pre-wrap; font-family:-apple-system, BlinkMacSystemFont,'Segoe UI',Roboto,'Noto Sans','Helvetica Neue',Arial;">
۱) افزایش ۳۰ تا ۳۴ درصدی بسته‌های اینترنت همراه (پایان ۱۴۰۲) : اپراتورها در دی‌ماه ۱۴۰۲ بسته‌های دیتای خود را ۳۰٪ تا ۳۴٪ افزایش دادند. این افزایش پس از چند سال تثبیت نسبی قیمت اینترنت رخ داد.

۲) درخواست اپراتورها برای افزایش ۵۰ تا ۷۵ درصدی تعرفه‌ها (۱۴۰۳–۱۴۰۴): اپراتورها به دلیل افزایش هزینه‌های ارزی، تجهیزات شبکه، انرژی و توسعهٔ زیرساخت، چند بار درخواست افزایش‌های سنگین‌تر داده‌اند. اگرچه همهٔ این درخواست‌ها تأیید نشده، اما نشان‌دهندهٔ شکاف شدید بین هزینهٔ واقعی ارائهٔ خدمات و تعرفه‌های موجود است.

۳) قیمت اینترنت ثابت (ADSL و فیبر): تعرفهٔ اینترنت ثابت معمولاً دیرتر از اینترنت همراه تغییر کرده است. در برخی سال‌ها قیمت اشتراک ثابت ماهانه تقریباً ثابت مانده و فقط قیمت «حجم اضافه» تغییراتی داشته است. با شروع توسعهٔ اینترنت فیبر نوری، برخی بسته‌ها افزایش قیمت داشته‌اند تا هزینه‌های توسعه شبکه جبران شود.
</pre>

## مقایسهٔ روند قیمت اینترنت با تورم عمومی
<pre style="white-space: pre-wrap; font-family:-apple-system, BlinkMacSystemFont,'Segoe UI',Roboto,'Noto Sans','Helvetica Neue',Arial;">
۱) دوره‌هایی که قیمت اینترنت کمتر از تورم رشد کرده: در سال‌هایی که تورم عمومی ایران بالاتر از ۳۰–۴۰٪ بوده، قیمت اینترنت ثابت مانده یا رشد بسیار کمی داشته است.
بنابراین در این دوره‌ها:
قیمت واقعی (real price) اینترنت برای کاربران عملاً کاهش یافته یعنی کاربران به ازای هر واحد پول، حجم بیشتری دیتا دریافت کرده‌اند. مثلاً در سال‌های ۱۳۹۸ تا ۱۴۰۱ تورم بالا بود اما قیمت بسته‌های عمده اینترنت تغییر چندانی نکرد. 

۲) دوره‌هایی که قیمت اینترنت بیشتر از تورم رشد کرده : در مقاطعی که تعرفه‌ها دچار «پرش ناگهانی» شدند، رشد قیمت اینترنت از تورم عمومی همان سال جلوتر بوده است.
نمونهٔ مشخص:
افزایش ۳۰–۳۴٪ پایان ۱۴۰۲
با توجه به اینکه تورم رسمی آن سال حدود ۴۰٪ بوده، این افزایش تقریباً نزدیک به تورم بود، اما با احتساب سال‌های قبل که قیمت‌ها ثابت مانده بودند، اثر آن برای کاربران محسوس‌تر بود. در صورتی که درخواست اپراتورها برای افزایش‌های ۵۰ تا ۷۵٪ تصویب شود، این افزایش از تورم عمومی همان سال جلوتر خواهد بود.
</pre>

## عوامل مؤثر بر افزایش قیمت اینترنت
<pre style="white-space: pre-wrap; font-family:-apple-system, BlinkMacSystemFont,'Segoe UI',Roboto,'Noto Sans','Helvetica Neue',Arial;">
۱) هزینه‌های ارزی و تجهیزات مخابراتی: تجهیزات شبکه (BTS, روتر، سوییچ، آنتن، سرور) عمدتاً وارداتی هستند
افزایش نرخ ارز = افزایش چند برابری هزینهٔ تأمین تجهیزات

۲) افزایش هزینهٔ انرژی و نگه‌داری: دیتاسنترها، سایت‌های موبایل و تجهیزات انتقال انرژی زیادی مصرف می‌کنند. افزایش قیمت انرژی بر هزینهٔ اپراتور تأثیر مستقیم دارد

۳) رشد شدید مصرف دیتا: میانگین مصرف دیتا نسبت به ده سال پیش چندین برابر شده. این رشد، توسعهٔ زیرساخت و سرمایه‌گذاری مداوم را لازم می‌کند

۴) سیاست‌های رگولاتوری: محدودیت یا تأخیر در تأیید افزایش تعرفه‌ها، الزام به ارائه بسته‌های حمایتی با قیمت پایین و اخذ عوارض از اپراتورها
</pre>

## جمع‌بندی نهایی
<pre style="white-space: pre-wrap; font-family:-apple-system, BlinkMacSystemFont,'Segoe UI',Roboto,'Noto Sans','Helvetica Neue',Arial;">
در ده سال اخیر:

1. تورم عمومی چندصد درصد رشد کرده و سطح قیمت‌ها چند برابر شده است.

2. قیمت اینترنت رشد یکنواخت نداشته و به شکل پله‌ای افزایش پیدا کرده است.

3. در بعضی بازه‌ها (۱۳۹۷–۱۴۰۱): قیمت اینترنت کمتر از تورم رشد کرده و عملاً ارزان‌تر شده است.

4. در برخی سال‌ها (۱۴۰۲–۱۴۰۳): افزایش‌های ۳۰ تا ۳۴ درصدی انجام شد. درخواست افزایش‌های ۵۰–۷۵ درصدی هم ارائه شده است. یعنی در این سال‌ها قیمت اینترنت در حال نزدیک شدن یا حتی سبقت گرفتن از تورم عمومی است.

5. فشار هزینه‌ای اپراتورها، نوسان ارز، توسعه زیرساخت و مصرف بالای دیتا، مهم‌ترین محرک‌های افزایش قیمت هستند.
</pre>
